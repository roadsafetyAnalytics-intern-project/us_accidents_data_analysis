```python
#Answer for the question 1 - How many unique values are there in categorical columns like City, State, and Weather_Condition? Which category is the most frequent in each?

print("Cities:", df['City'].nunique())
print("Most frequent City:", df['City'].value_counts().idxmax())

print("States:", df['State'].nunique())
print("List of States:", df['State'].unique())
print("Most frequent State:", df['State'].value_counts().idxmax())

print("Weather Conditions:", df['Weather_Condition'].nunique())
print("List of Weather Conditions:", df['Weather_Condition'].unique())
print("Most frequent Weather Condition:", df['Weather_Condition'].value_counts().idxmax())
```

```python
# Answer for the question 2 - Calculate the percentage of missing values in each column and list columns with more than 5% missing data.

missing_percent = (df.isnull().sum() / len(df)) * 100
print("Columns with more than 5% missing values:\n")
print(missing_percent[missing_percent > 5])
```

```python
# Answer for the question 3 - Convert Start_Time and End_Time columns to datetime objects. 
# Find the range (earliest and latest) of accident start times in the dataset.

df['Start_Time'] = pd.to_datetime(df['Start_Time'], errors='coerce', format='mixed')
df['End_Time'] = pd.to_datetime(df['End_Time'], errors='coerce', format='mixed')

print("Earliest Accident Start Time:", df['Start_Time'].min())
print("Latest Accident Start Time:", df['Start_Time'].max())
```

```python
# Answer for the question 4 - Identify any duplicate records in the dataset based on all columns. 
# How many duplicates exist and how will you handle them?

print("Number of duplicate records:", df.duplicated().sum())

df = df.drop_duplicates()
print("Number of duplicate records after removal:", df.duplicated().sum())
```

```python
# Answer for the question 5 - Explore the distribution of accident severity (Severity column). 
# What percentage of accidents belong to each severity level? Visualize this distribution with a pie chart.

import matplotlib.pyplot as plt

severity_counts = df['Severity'].value_counts(normalize=True) * 100
print("Percentage of accidents by Severity level:\n", severity_counts)

severity_counts.plot.pie(autopct='%1.1f%%', figsize=(6,6))
plt.ylabel("")
plt.title("Accident Severity Distribution")
plt.show()
```
